```python
from typing import Any, Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class BobState(TypedDict):
    name: str


def CongratulateBob(state: BobState) -> dict:
    return {"name": f" {state['name']}! You have successfully completed the task."}
```

```python
graph = StateGraph(BobState)
graph.add_node("start", CongratulateBob)
graph.set_entry_point("start")
graph.set_finish_point("start")
app = graph.compile()
```

```python
from IPython.display import display, Image
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"name": "Bob"})
```
